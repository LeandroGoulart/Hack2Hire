## 1️⃣ PROPOSTA 1: Amazon Bedrock + BDA (Iniciante)

### 🎯 Descrição
- **Caso de uso:** Fintech CrediFácil (empréstimos imobiliários)
- **Mentores:** Layrane Dantas, Lucas Leite, Gilson Castro
- **Duração:** Não especificada (provavelmente 8+ horas)
- **Stack:** Python + boto3 + Amazon Bedrock Data Automation (BDA) + Amazon Nova

### 📋 Escopo
- Receber pacote de documentos (5-8 PDFs por solicitação)
- Classificar automaticamente cada tipo de documento
- Extrair dados relevantes
- Retornar JSON estruturado e padronizado
- **Pipeline simples:** upload → extração → estruturação

### 🔧 Serviços AWS Utilizados
| Serviço | Função |
|---------|--------|
| **Amazon S3** | Armazenamento de documentos |
| **Amazon Bedrock Data Automation (BDA)** | Classificação e extração automática ⭐ |
| **Amazon Nova** | Estruturação de dados (tool-calling) |
| **AWS Lambda** | Automação do fluxo |
| **AWS Step Functions** | Orquestração |
| **AWS IAM** | Permissões |

### ✅ Vantagens

| Vantagem | Detalhes |
|----------|----------|
| **Simplificado** | BDA faz o trabalho pesado de extração e classificação |
| **Menos código** | Menos integração entre serviços = menos pontos de falha |
| **Documentação clara** | Foco específico em BDA com exemplos diretos |
| **Time menor** | 2-3 pessoas conseguem completar |
| **Prototipagem rápida** | MVP funcional em poucas horas |
| **Menos serviços** | Não requer Textract, Comprehend, ou Step Functions complexos |

### ❌ Desvantagens

| Desvantagem | Detalhes |
|------------|----------|
| **BDA é novo** | Lançado recentemente, menos exemplos na comunidade |
| **Escopo limitado** | Sem agente conversacional ou RAG |
| **Não escalável** | Difícil adicionar funcionalidades depois (Q&A, histórico) |
| **Sem busca semântica** | Não há mecanismo de consulta de documentos |
| **Obrigatório usar BDA** | Limite técnico: não permite OCR próprio |
| **Sem interface com usuário** | Apenas JSON, sem visualização ou dashboard |
| **Menos "wow factor"** | Menos inovador para apresentação final |

### ⏱️ Cronograma Recomendado
- **Setup (30 min):** Criar S3, habilitar BDA, IAM
- **Implementação (2-3h):** Criar projeto BDA, pipeline, Lambda

## 💻 COMPARAÇÃO DE CÓDIGO

### Proposta 1: BDA (Mais alto nível, menos código)
```python
# Usar BDA para fazer o trabalho pesado
bda_client = boto3.client('bedrock-data-automation')
project = bda_client.create_data_automation_project(
    project_name='emprestimos',
    blueprint='document_classification'
)
# BDA cuida de tudo: classificação + extração
result = bda_client.start_document_classification_job(
    document_s3_path='s3://bucket/documents/'
)
```

### Proposta 2: Orquestração Complexa (Múltiplos serviços)
```python
# Você coordena cada peça
textract = boto3.client('textract')
comprehend = boto3.client('comprehend')
bedrock = boto3.client('bedrock-runtime')

# Textract extrai
text = textract.detect_document_text(...)
# Comprehend classifica
classification = comprehend.classify_document(...)
# Bedrock normaliza
normalized = bedrock.invoke_model(...)
# Step Functions orquestra tudo
stepfunctions.start_execution(...)
```

### Proposta 3: Agente Automático (Framework cuida da orquestração)
```python
from strands import Agent
from strands.models import BedrockModel

# Você define ferramentas, agente decide a ordem
@tool
def extrair(bucket, key): return textract.detect_document_text(...)

@tool
def salvar(resultado): return dynamodb.put_item(resultado)

# Agente coordena tudo automaticamente
agent = Agent(
    model=BedrockModel('amazon-nova-pro'),
    tools=[extrair, salvar],
    system_prompt='Você é um analista de sinistros...'
)

resultado = agent("Analise o documento no bucket X")
```
